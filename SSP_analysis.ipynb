```python
#k-mer
from collections import Counter
import numpy as np

def kmer_profile(seqs, k=6):
    kmers = Counter()
    for seq in seqs:
        for i in range(len(seq) - k + 1):
            kmer = seq[i:i+k]
            kmers[kmer] += 1
    return kmers

def normalize_counter(counter, all_keys):
    return np.array([counter.get(k, 0) for k in all_keys])

# K-mer профили
wt_kmers = kmer_profile(reads_wt, k=6)
mut1_kmers = kmer_profile(reads_mut1, k=6)
mut2_kmers = kmer_profile(reads_mut2, k=6)

# Объединённое множество всех уникальных k-mers
all_kmers = set(wt_kmers) | set(mut1_kmers) | set(mut2_kmers)

# Векторы
wt_vec = normalize_counter(wt_kmers, all_kmers)
mut1_vec = normalize_counter(mut1_kmers, all_kmers)
mut2_vec = normalize_counter(mut2_kmers, all_kmers)
```

```python
from Bio.Align import PairwiseAligner

def align_and_get_mutations(wt_seq, mut_seq):
    aligner = PairwiseAligner()
    aligner.match_score = 2
    aligner.mismatch_score = -1
    aligner.open_gap_score = -2
    aligner.extend_gap_score = -0.5
    # The align method can return multiple alignments, we'll take the first one
    alignments = aligner.align(wt_seq, mut_seq)
    if not alignments:
        return [] # Return empty list if no alignments found

    alignment = alignments[0]

    # Access aligned sequences correctly using the sequences attribute
    aligned_wt = alignment.sequences[0]
    aligned_mut = alignment.sequences[1]

    mutations = []
    # Iterate through the aligned sequences to find differences
    for i, (w, m) in enumerate(zip(aligned_wt, aligned_mut)):
        if w != m and w != '-' and m != '-':
            # Store mutation as (position in alignment, ref base, alt base)
            mutations.append((i, w, m))
        # Note: Insertions and deletions (indels) are represented by '-'
        # If you want to capture indels, you would add conditions for w == '-' or m == '-'
    return mutations
```

The genetic distance calculated using k-mer frequency vectors specifically measures how different the sequences of the mutants are compared to the Wild Type at a sequence level. It quantifies the dissimilarity in their k-mer composition.

The average fitness (AUC), which is calculated from your OD.xlsx data, represents the growth or metabolic activity over time for each group.

So, the fitness landscape plots you see are essentially showing:

X-axis: How genetically different each mutant is from the WT (based on sequence).
Y-axis: How well each mutant (and WT) grew in a specific condition (based on AUC).

```python
from scipy.spatial.distance import cosine
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Assuming fitness_broth and fitness_antibiotic dictionaries and mut1_vec, mut2_vec, wt_vec are already defined

# Calculate genetic distance from WT (WT is the 0 point)
genetic_distance_from_wt = {
    "WT": 0.0,
    "Mut1": cosine(wt_vec, mut1_vec),
    "Mut2": cosine(wt_vec, mut2_vec)
}

print("Genetic Distance from WT (cosine):", genetic_distance_from_wt)

# Calculate average fitness (AUC) for each group
fitness_broth = {
    "WT": df.loc[df['Sample'].isin(['1A', '1B', '1C']), 'AUC'].mean(),
    "Mut1": df.loc[df['Sample'].isin(['2A', '2B', '2C']), 'AUC'].mean(),
    "Mut2": df.loc[df['Sample'].isin(['3A', '3B', '3C']), 'AUC'].mean()
}

fitness_antibiotic = {
    "WT": df.loc[df['Sample'].isin(['5A', '6A', '7A']), 'AUC'].mean(),
    "Mut1": df.loc[df['Sample'].isin(['5B', '6B', '7B']), 'AUC'].mean(),
    "Mut2": df.loc[df['Sample'].isin(['5C', '6C', '7C']), 'AUC'].mean()
}


# Function to plot fitness landscape with WT as 0 point
def plot_fitness_landscape_from_wt(fitness, genetic_distance, condition):
    x = [genetic_distance["WT"], genetic_distance["Mut1"], genetic_distance["Mut2"]]
    y = [fitness["WT"], fitness["Mut1"], fitness["Mut2"]]
    labels = ["WT", "Mut1", "Mut2"]

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, 'o-', color='slateblue')
    for i in range(3):
        # Adjust text position based on the label and condition if needed, or a general offset
        plt.text(x[i] + 0.001, y[i] + 0.5, labels[i], fontsize=12) # Adjusted text position again

    plt.xlabel("Genetic Distance from WT (cosine)")
    plt.ylabel("Average Fitness (AUC)")
    plt.title(f"Fitness Landscape from WT ({condition})")
    plt.grid(True)
    plt.show()

# Plot for Broth only
plot_fitness_landscape_from_wt(fitness_broth, genetic_distance_from_wt, "Broth only")

# Plot for Broth + Antibiotic
plot_fitness_landscape_from_wt(fitness_antibiotic, genetic_distance_from_wt, "Broth + Antibiotic")
```

```python
import scipy.stats as stats
import pandas as pd

# Assuming df is already loaded with 'Sample' and 'AUC' columns

# Define the groups
wt_samples = ["1A", "2A", "3A"]
mut1_samples = ["1B", "2B", "3B"]
mut2_samples = ["1C", "2C", "3C"]

wt_ab_samples = ["5A", "6A", "7A"]
mut1_ab_samples = ["5B", "6B", "7B"]
mut2_ab_samples = ["5C", "6C", "7C"]

# Extract AUC values for each group in Broth only condition
wt_broth_auc = df[df["Sample"].isin(wt_samples)]["AUC"].dropna()
mut1_broth_auc = df[df["Sample"].isin(mut1_samples)]["AUC"].dropna()
mut2_broth_auc = df[df["Sample"].isin(mut2_samples)]["AUC"].dropna()

# Perform one-way ANOVA for Broth only
print("--- ANOVA for Broth only condition ---")
if len(wt_broth_auc) > 0 and len(mut1_broth_auc) > 0 and len(mut2_broth_auc) > 0:
    f_statistic_broth, p_value_broth = stats.f_oneway(wt_broth_auc, mut1_broth_auc, mut2_broth_auc)
    print(f"F-statistic: {f_statistic_broth:.4f}")
    print(f"P-value: {p_value_broth:.4f}")
else:
    print("Insufficient data for ANOVA in Broth only condition.")

print("\n--- ANOVA for Broth + Antibiotic condition ---")
# Extract AUC values for each group in Broth + Antibiotic condition
wt_ab_auc = df[df["Sample"].isin(wt_ab_samples)]["AUC"].dropna()
mut1_ab_auc = df[df["Sample"].isin(mut1_ab_samples)]["AUC"].dropna()
mut2_ab_auc = df[df["Sample"].isin(mut2_ab_samples)]["AUC"].dropna()


# Perform one-way ANOVA for Broth + Antibiotic
if len(wt_ab_auc) > 0 and len(mut1_ab_auc) > 0 and len(mut2_ab_auc) > 0:
    f_statistic_ab, p_value_ab = stats.f_oneway(wt_ab_auc, mut1_ab_auc, mut2_ab_auc)
    print(f"F-statistic: {f_statistic_ab:.4f}")
    print(f"P-value: {p_value_ab:.4f}")
else:
     print("Insufficient data for ANOVA in Broth + Antibiotic condition.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import numpy as np

# Combine AUC values for Broth only
broth_auc_combined = np.concatenate([wt_broth_auc, mut1_broth_auc, mut2_broth_auc])

# Create corresponding labels
broth_labels = ["WT"] * len(wt_broth_auc) + ["Mutant1"] * len(mut1_broth_auc) + ["Mutant2"] * len(mut2_broth_auc)
```

```python
# Perform Tukey's HSD test
tukey_results_broth = pairwise_tukeyhsd(endog=broth_auc_combined, groups=broth_labels, alpha=0.05)

# Print the results
print("--- Tukey's HSD Post-hoc Test Results for Broth only condition ---")
print(tukey_results_broth)
```

```python
def categorize_mutation(row):
    ref = row['Reference']
    alt = row['Alternate']

    # Check if it's a substitution (not an indel)
    if ref != '-' and alt != '-':
        # Define purines and pyrimidines
        purines = {'A', 'G'}
        pyrimidines = {'C', 'T'}

        if ref in purines and alt in purines:
            return 'Transition'
        elif ref in pyrimidines and alt in pyrimidines:
            return 'Transition'
        elif ref in purines and alt in pyrimidines:
            return 'Transversion'
        elif ref in pyrimidines and alt in purines:
            return 'Transversion'
    # If you had indels, you could add conditions here, e.g.,
    # elif ref == '-' or alt == '-':
    #     return 'Indel'
    return 'Other' # For cases not covered (e.g., N in sequence, or if indels were present but not explicitly handled)

# Apply the categorization function to both DataFrames
df_mutations_mut1['Mutation Type'] = df_mutations_mut1.apply(categorize_mutation, axis=1)
df_mutations_mut2['Mutation Type'] = df_mutations_mut2.apply(categorize_mutation, axis=1)

# Calculate the frequency of each mutation type
mutation_frequencies_mut1 = df_mutations_mut1['Mutation Type'].value_counts().to_dict()
mutation_frequencies_mut2 = df_mutations_mut2['Mutation Type'].value_counts().to_dict()

print("Mutation Frequencies for Mutant 1 vs WT:")
print(mutation_frequencies_mut1)

print("\nMutation Frequencies for Mutant 2 vs WT:")
print(mutation_frequencies_mut2)
```
